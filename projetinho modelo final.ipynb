#### *Primeiro passo* : 

```python
from scipy.stat import linregress
```

```python
arquivo = pd.read_csv("C:/Users/davi-/OneDrive/Documentos/Arquivos/Users/Materiais_Assuntos/Ciência de dados/projeto ciência de dados/def_area_2004_20191.csv")
arquivo.rename(columns={'1':'ANO_ESTADO','1.1':'AC','1.2':'AM','1.3':'AP','1.4':'MA','1.5':'MT','1.6':'PA','1.7':'RO','1.8':'RR','1.9':'TO','1.10':'AMZ_LEGAL'},inplace=True)
```

```python
arquivo
```

```python
result = linregress(arquivo["ANO_ESTADO"], arquivo["TO"])
```

```python
a = result.slope
```

```python
b = result.intercept
```

```python
x = arquivo["ANO_ESTADO"]
```

```python
y = arquivo["TO"]
```

```python
cont = y.count()
```

```python
import matplotlib.pyplot as plt

def gen_line(x_min, x_max, n_points, a, b):
    interval = x_max-x_min
    interval_value = interval/n_points
    
    x_result = []
    y_result = []

    for i in range(n_points):
        current_x = x_min+i*interval_value
        x_result.append(current_x)
        y_result.append(a*current_x + b)
        
    return x_result, y_result

def show_regression(x, y, a, b):
    N = 100

    plt.scatter(x, y)

    line_1 = gen_line(min(x), max(x), N, a, b)
    plt.plot(line_1[0], line_1[1], c= "red")

    plt.gca()
    plt.show()
```

```python
gen_line(x[0],x[15],cont,a,b)
```

```python
show_regression(x, y, a, b)
```

```python
def predict(x):
    y_predict = b + a*x
    
    return y_predict
```

```python
predict(2019)
```

```python
predict(2020)
```
